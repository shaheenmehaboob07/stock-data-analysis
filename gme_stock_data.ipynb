# Use yfinance to Extract GameStop Stock Data

This notebook extracts historical GameStop (GME) stock data using yfinance, resets the index, and displays the first five rows.

```
import yfinance as yf
```

```
gme_data = yf.Ticker('GME')
gme_data = gme_data.history(period='max')
gme_data.reset_index(inplace=True)
gme_data.head()
```
